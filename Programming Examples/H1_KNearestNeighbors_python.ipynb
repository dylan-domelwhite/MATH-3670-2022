# K-Nearest Neighbors Example

In this example, we create a training set and use it to find the K-nearest neighbors regression function. First, we generate a training set of $N = 1000$ points $\{(x_n, y_n)\}_{n=1}^N$ where $x_n$ are independent and uniformly distributed in $[0, 2]$ and $y_n = 2x_n + \sin(2\pi x_n) + \varepsilon_n$ where $\varepsilon_n$ are independent and normally distributed with mean $0$ and standard deviation $\sigma = 1$.

```python
from matplotlib import pyplot as plt
import numpy as np

N = 1000
sigma = 1

x_train = 2*np.random.rand(N)
eps = np.random.normal(0, sigma, N)
y_train = 2*x_train + np.sin(2*np.pi*x_train) + eps;
```

We plot the training set in green and the theoretical regression function in blue.

```python
x = np.linspace(0, 2, 100)
y = 2*x + np.sin(2*np.pi*x)

fig, ax = plt.subplots()
ax.set_title('Training Data')
plt.scatter(x_train,y_train, c='g')
plt.plot(x, y, c='b')
plt.show()
```

Next, we use the *sklearn* package to find the K-nearest neighbor regression function. Then, we plot the theoretical regression function and the K-nearest neighbor regression function in red. Feel free to play with the parameters to see how the nearest neighbor regression changes.

```python
from sklearn.neighbors import KNeighborsRegressor

K = 25

model = KNeighborsRegressor(n_neighbors = K)
model.fit(x_train.reshape(-1, 1), y_train)


y_predicted = model.predict(x.reshape(-1,1))

fig, ax = plt.subplots()
ax.set_title('{}-Nearest Neighbor Regression Function'.format(K))
plt.plot(x, y, c='b')
ax.plot(x, y_predicted, 'r')
plt.show()
```
